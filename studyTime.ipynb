```python
# A simple script to calculate BMI
#from pywebio.input import input, input_group, FLOAT, file_upload, select, radio
#from pywebio.output import put_text, put_image, put_warning, put_success
from pywebio.input import *
from pywebio.output import *
import pandas as pd

# following 4 commands (+another 2) necessary to make webpage reloadable
from flask import Flask
from pywebio.platform.flask import webio_view
from pywebio import STATIC_PATH
app = Flask(__name__) # create a flask app?

def studyTime():
    
    df = pd.read_csv(filepath_or_buffer='few_grades.csv', sep=',')
    mean_age = df['Number'].mean(axis=0)
    
    #Checking that we can output data from uploaded files here. 
    #put_text('The value you got from csv file is: %.1f.' % (mean_age))
    style(put_text('The value you got from csv file is: %.1f.' % (mean_age)), "color: green")
    
    #img = open('gmail_logo.png', 'rb').read()  
    #put_image(img, width='50px')
    
    put_image('https://www.python.org/static/img/python-logo.png', width='150px')
    
    user_info = input_group("User info",[
      input('Input your name', name='name', required=True),
      input('Input your password', name='password', type=PASSWORD, required=True)
    ])
    
    # temporary password identification
    if user_info['password'] != 'right':
        put_warning('wrong password. Refresh to input again.')
    else:
        put_success('Welcome to StudyTime, %s' % user_info['name'])
    
        major = select("What Major are you in?", options=[
            'Computer Science', 'Data Science', 'Applied Mathematics', 'Chemistry'])
        
        if major != 'Computer Science':
            put_warning('This category is not ready yet.')
        else:
            put_success('You are a %s student' % (major))
            
            course = radio("Which course do you want to inquire about?", options=[
                'CST1100', 'CST1101', 'CST2402', 'CST3512'])
            
            # checking the input gotten for course is correct
            put_text('The course you take is: %s.' % (course))
            
            # a pywebio function to upload in the website
            #uploaded_file = file_upload(label='Upload your CSV file', accept='.csv')
    
            survey_answers = input_group("User info",[
                input('What grade do you want in this class? A, B, C?', name='desired_grade', type="text", required=True),
                #type=STRING or type=TEXT does not work here
                input('How many hours do you spend studying for this class?', name='hours_spent', type=FLOAT, required=True),
                input('How many hours can you study per week?', name='hours_available', type=FLOAT, required=True),
                input('Do you have an upcoming exam for this class? Yes|No?', name='exam', type="text"),
                input('Would you like an alarm to remind you a week before your exam? Yes|No', name='alarm', type="text") 
                # how to send alarm?
            ])
    
            previous_average = df.loc[df['course_number'] == course, 'Letter'].item() # instead of .item(), maybe .iloc[0]
            
            put_text('The previous average of your course %s is %s' % (course, previous_average))
        
            recommended_hours = {"CST1100":{"A":6, "B":4, "C":3}, 
                                 "CST1101":{"A":6, "B":4, "C":3}, 
                                 "CST2402":{"A":8, "B":6, "C":4}, 
                                 "CST3512":{"A":12, "B":10, "C":8}} # should b different 4 each course depending on what previous_average was 
            # maybe use a dictionaries within a dictionary?
            
            now_recommended_hours = recommended_hours[course][survey_answers['desired_grade']]
            
            if survey_answers['hours_spent'] >= now_recommended_hours:
                put_text('The number of hours recommended is %.1f. You are already studying enough for this course.' % (now_recommended_hours))
            else:
                put_text('The number of hours recommended is %.1f. You should study an additional %.1f hours.' % (now_recommended_hours, now_recommended_hours - survey_answers['hours_spent']))

# following 2 commands also necessary to make a reloadable webpage
app.add_url_rule('/', 'webio_view', webio_view(studyTime), methods=['GET','POST','OPTIONS'])
app.run(host="localhost", port=80)

# The following 2 commands don't seem necessary if the previous 2 are ran. not entirely sure what consequences there'd b
if __name__ == '__main__':
    studyTime()
```
